### 2.1 Variables & expressions
A **variable** is a name that refers to a value. You create one with the assignment operator `=`.
Variable names can be any length but must **start with a letter or underscore**, followed by letters, numbers,
or underscores and can't be a reserved keyword.

````python
# valid assignments    <---- a sentence starting with a hash tag is a comment and is never executed.
x = 5               # define an variable just like what we once learn in Math class
print(x)
````

````python
# multiple assignment at once
x = y = z = 5         # all three get 5
print(x, y, z)
````

````python
a, b = 3, 4           # 'unpacking': a=3, b=4
print(a, b)
````

````python
# use an underscore as thousands separator 1,000,000
my_wealth = 1_000_000
print(my_wealth)
````

````python
# These are INVALID
# 1_model = 2     # can't start with a digit
# return = 3      # 'return' is a reserved keyword
````

````python
# addition/ plus
a_plus_b = a + b
print(a_plus_b)
````

````python
# subtaction/ minus
a_minus_b = a - b
print(a_minus_b)
````

````python
# multiplication/ times
a_times_b = a * b
print(a_times_b)
````

````python
# division/ divided by
a_over_b = a / b
print(a_over_b)
````

### 2.2 Booleans
A **boolean** is either `True` or `False`. Booleans power every decision your code makes. Check a value's type
with `isinstance(value, type)`.

````python
exam_pass = True
print(exam_pass)
print(type(exam_pass))
print(isinstance(exam_pass, bool))   # True
````

Compound conditions use `and`, `or`, `not`. Python **short-circuits**: in `A and B`, if `A`
is `False`, `B` is never evaluated.

````python
exam_pass = True
graduation = False

print(exam_pass and graduation)
print(exam_pass or graduation)
print(not exam_pass)
print(not graduation)
````

### 2.3 Numbers: `int` and `float`
- **`int`** — whole numbers (`21`, `-4`, `1000000`).
- **`float`** — numbers with a decimal point (`0.95`, `-3.14`).

````python
epoch = 21
pi = 3.1415926
print(epoch, type(epoch))
print(pi, type(pi))
print('is epoch an int? ', isinstance(epoch, int))
print('is pi a float?   ', isinstance(pi, float))
````

````python
# round the demcimal to certain digit: round(a_float, ndigits)
pi_rounded = round(pi, 2)
print(pi_rounded)
````

### 2.4 Strings
A **string** (`str`) is text — a sequence of characters in single or double quotes. Strings are **indexable**
(0-based) and **sliceable**, but **immutable** (you can't change a character in place).

<p align="left">
  <img src="https://raw.githubusercontent.com/cgpan/humd5001_mini/main/figures/string_slice.png" width="700">
</p>

````python
status = 'success'
print(type(status))
print('first char :', status[0])      # 's'
print('first two  :', status[:2])     # 'su'
print('last two   :', status[-2:])    # 'ss'
print('length     :', len(status))    # using len() to show the total length
````

````python
# Handy string methods
print(status.upper())           #
print(status.upper().lower())   # You can attach two or more functions, but it is rare.
print(status.startswith('a'))   # False
print(status.endswith('s'))     # True
print('  trim me  '.strip())    # remove surrounding whitespace
````

**Immutability** — this fails because strings can't be edited in place:

````python
status = 'success'
# status[0] = 'X'        # TypeError: 'str' object does not support item assignment
status = 'Xuccess'        # but REASSIGNING the whole variable is fine
print(status)
````

### 2.5 Lists — the workhorse container
A **list** is an *ordered, mutable* collection written in `[ ]`. It can hold mixed types, be indexed and
sliced, grown, and looped over.

<p align="left">
  <img src="https://raw.githubusercontent.com/cgpan/humd5001_mini/main/figures/string_slice.png" width="700">
</p>

*Image Source:* [scientecheasy](https://www.scientecheasy.com/2023/03/python-slice-list.html/#google_vignette)

````python
financial_derivatives = ['futures', 'swaps', 'options']
print(type(financial_derivatives))
print('first      :', financial_derivatives[0])
print('first 3 chars of first item:', financial_derivatives[0][:3])

# you can count the number of elements in the list
deri_count = len(financial_derivatives)
print('The number of the current finanacial derivatives is {}'. format(deri_count))

# lists are MUTABLE: change, append, extend
financial_derivatives[1] = 'forwards'      # change an element
financial_derivatives.append('bonds')      # add to the end
print(financial_derivatives)
````

````python
# string can be converted into list --> using .split()
got = "You know nothing, Jon Snow."
# split the string based on the space
got_vocab = got.split()
print(got_vocab)
````

````python
# split the string based on the punctuation like comma or period
got_split = got.split(",")
print(got_split)
````

````python
# Lists can be concatenated with +
variables = ['x1', 'x1', 'x2'] + ['x3', 'x4']
print(variables)
````

````python
# you can remove any unit one at a time.
variables.remove("x1")
````

### 2.6 Tuples — like lists, but frozen
A **tuple** is an *ordered, **immutable*** collection, written with `( )` (or just commas). Use tuples for
fixed groupings that shouldn't change.

````python
coffees = ('brazilian', 'costa_rican', 'colombian')
print(type(coffees), '| first:', coffees[0])

# parentheses are optional — commas make the tuple
# But I recommend adding the parentheses
grab_bag = 4, 'EHR', ['year', 'in', 'review']
print(grab_bag)

# coffees[0] = 'kenyan'    # TypeError: tuples are immutable
````

### 2.7 Sets — unique, unordered
A **set** holds *unique*, *unordered* values in `{ }`. Great for de-duplicating and for fast membership tests
with `in`. (Note: `{}` alone makes an empty **dict**, not a set — use `set()`.)

````python
peanuts = {'snoopy', 'snoopy', 'woodstock'}
print(peanuts)                      # duplicates removed automatically
print('snoopy' in peanuts)          # membership test -> True

set1 = {'python', 'R'}
set2 = {'R', 'julia'}
print('union       :', set1.union(set2))           # everything
print('intersection:', set1.intersection(set2))    # in both
````

### 2.8 Dictionaries — key → value lookups
A **dict** maps **keys** to **values** in `{key: value}`. It's the most important container for data work
(think: column name → column, or id → record).

````python
airports = {'LAGUARDIA': 'LGA', 'LOGAN': 'BOS', 'NANTUCKET': 'ACK'}
print(airports['LOGAN'])                 # look up by key -> 'BOS'
print(airports.get('KENNEDY'))           # .get() returns None instead of erroring on a missing key

airports['DULLES'] = 'IAD'               # add / update a pair
print(airports.keys())
print(airports.values())
````

````python
# Loop over keys and values together with .items()
for name, code in airports.items():
    print(f'{name:12s} -> {code}')
````

### 2.9 Ranges and type conversions
**`range(start, stop, step)`** generates a sequence of integers (stop is *exclusive*) — most often used to drive
loops. **Type conversions** use the type's name as a function: `int()`, `float()`, `str()`, `list()`.

````python
print(range(1,10,2))
print(list(range(1,10,2)))
````

## 3. Conclusions & Key Takeaways
- **Variables** name values; check types with `type()` / `isinstance()`.
- **Scalars:** `bool`, `int`, `float`, `str`. **Containers:** `list`, `tuple`, `set`, `dict`, plus `range`.
- **Mutable** (changeable in place): `list`, `set`, `dict`. **Immutable**: `int`, `float`, `bool`, `str`, `tuple`.
- Strings and lists are **0-indexed** and support slicing `[start:stop]`; `stop` is exclusive.
- Use **f-strings** (`f'{x}'`) for readable string building.
- **Dicts** (key→value) and **lists** are the two containers you'll use most in data science.
- Conversions can fail — e.g. `int('3.8')` errors; go `int(float('3.8'))` instead.

### 🧾 Quick Reference
| Type | Example | Mutable? | Ordered? |
|---|---|:--:|:--:|
| `bool` | `True` | – | – |
| `int` / `float` | `21` / `0.95` | no | – |
| `str` | `'success'` | **no** | yes |
| `list` | `['a', 5]` | **yes** | yes |
| `tuple` | `('a', 5)` | no | yes |
| `set` | `{'a', 'b'}` | yes | no |
| `dict` | `{'k': 'v'}` | yes | yes (insertion) |
